Model Training And Testing

```python
df=df.reset_index(drop=True)
df
```

```python
df.columns
```

```python
df2=df.drop(['Artist Name','Track Name','mode','time_signature','Class','Collabration'],axis='columns')
df2
```

```python
def normalize(feature):
    return (feature-np.mean(feature))/np.std(feature)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
df2=scaler.fit_transform(df2)
df2
```

```python
df.columns
```

```python
df2=pd.DataFrame(df2,columns=['Popularity','danceability', 'energy','key', 'loudness', 'speechiness', 'acousticness','instrumentalness', 'liveness', 'valence', 'tempo','duration_in min/ms'])
df2
```

```python
df=df.drop(['Popularity','danceability', 'energy','key', 'loudness', 'speechiness', 'acousticness','instrumentalness', 'liveness', 'valence', 'tempo','duration_in min/ms'],axis='columns')
```

```python
df
```

```python
df=pd.concat([df,df2],axis='columns')
df
```

```python
list(df['Artist Name'].unique())
```

```python
len(df['Artist Name'].unique())
```

```python
df=df.drop(['Artist Name','Track Name'],axis='columns')
df
```

```python
Y=df['Class']
Y
```

```python
X=df.drop(['Class'],axis='columns')
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=77)
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(multi_class='multinomial')

lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

Standard Vector Machines(SVC)

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,Y_train)
```

```python
svc.score(X_test,Y_test)
```

Random forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
rfc.score(X_test,Y_test)
```

K Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```

XGBoost Classifier

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,Y_train);
```

```python
xgb.score(X_test,Y_test)
```

Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
gnb.score(X_test,Y_test)
```

```python
from sklearn.model_selection import cross_val_score
logistic=cross_val_score(lr,X,Y,cv=10)
np.average(logistic)
```

```python
svm=cross_val_score(svc,X,Y,cv=10)
np.average(svm)
```

```python
rfC=cross_val_score(rfc,X,Y,cv=10)
np.average(rfC)
```

```python
Knn=cross_val_score(knn,X,Y,cv=10)
np.average(Knn)
```

```python
Xgb=cross_val_score(xgb,X,Y,cv=10)
np.average(Xgb)
```

```python
Gnb=cross_val_score(gnb,X,Y,cv=10)
np.average(Gnb)
```

Evaluation of Metric

```python
from sklearn.model_selection import cross_val_score

models=[lr,svc,rfc,knn,xgb,gnb]
accuracies=[]

for i in models:
    score=cross_val_score(i,X,Y,cv=5)
    accuracies.append(np.mean(score))
    
```

```python
accuracies
```

```python
from sklearn import metrics

for i in models:
    Y_predicted=i.predict(X_test)
    matrix=metrics.confusion_matrix(Y_predicted,Y_test)
    sns.heatmap(matrix,annot=True)
    plt.show()
```
